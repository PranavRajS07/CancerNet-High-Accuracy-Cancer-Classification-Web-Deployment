**prediction for brain tumor**

```python
from tensorflow.keras.preprocessing import image
import numpy as np
from tensorflow.keras.models import load_model
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

model = load_model('/content/brain_50epoch.h5')

class_mapping = {
    0: 'brain menin',
    1: 'brain tumor',
    2: 'Class_2',
    3: 'Class_3'
    }

def load_and_preprocess_image(img_path):
    img = image.load_img(img_path, target_size=(224, 224))
    img_array = image.img_to_array(img)
    img_array = np.expand_dims(img_array, axis=0)
    img_array /= 255.0
    return img_array

def predict_and_display_image(img_path, model):
    img_array = load_and_preprocess_image(img_path)

    predictions = model.predict(img_array)
    predicted_class_index = np.argmax(predictions)

    predicted_class = class_mapping.get(predicted_class_index, 'Unknown Class')

    img = mpimg.imread(img_path)
    imgplot = plt.imshow(img)
    plt.title(f"Predicted Class: {predicted_class}")
    plt.show()

image_path_to_predict = '/content/brain_tumor_0067.jpg'

predict_and_display_image(image_path_to_predict, model)
```
